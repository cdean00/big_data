```python
# Use this cell to import your properly formatted data file 
# Column headers and data only- look at the whole file!

import pandas as pd

df = pd.DataFrame.from_csv('./datasets/data_formatted.txt', index_col=None)
```

```python
# Explore data

df.columns
```

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
mpl.rcParams['figure.figsize'] = (10.0, 10.0)
```

```python
# Use this cell to plot things

plt.figure(1)
plt.subplot(221)
plt.plot(df['millisecs'], df['pressure(Pascal)'])
plt.legend(loc='lower right')

plt.subplot(222)
plt.plot(df['millisecs'], df['temp_external'])
plt.legend(loc='lower right')

plt.subplot(223)
plt.plot(df['millisecs'], df['x_accel'], 'b-', label='X Accel')
plt.plot(df['millisecs'], df['y_accel'], 'r-', label='Y Accel')
plt.plot(df['millisecs'], df['z_accel'], 'g-', label='Z Accel')
plt.legend(loc='upper right')

plt.subplot(224)
plt.plot(df['millisecs'], df['x_mag'], 'b-', label='X Mag')
plt.plot(df['millisecs'], df['y_mag'], 'r-', label='Y Mag')
plt.plot(df['millisecs'], df['z_mag'], 'g-', label='Z Mag')
plt.legend(loc='upper right')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(2)
ax = fig.add_subplot(111, projection='3d')

ax.plot(df['millisecs'],df['x_accel'], df['y_accel'])
```
